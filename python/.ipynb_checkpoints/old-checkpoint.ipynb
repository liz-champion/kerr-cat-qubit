```python
import numpy as np
import matplotlib.pyplot as plt

import qutip as qt
```

```python
squeeze_drive = True
x_drive = True

arg_e_x = 0. #0.5 * np.pi

N = 10 # 10-d Hilbert space
MHz = 1. # work in units of MHz
hbar = 1. # work in units where hbar = 1
K = 2. * np.pi * 6.7 * MHz # Kerr nonlinearity
g_3 = 2. * np.pi * 20.0 * MHz # third-order nonlinearity
e_2 = 2. * np.pi * 15.5 * MHz if squeeze_drive else 0. # squeezing drive strength
xi = e_2 / (3. * g_3)
e_x = np.exp(1.j * arg_e_x) * 2. * np.pi * 740.0e-3 * MHz if x_drive else 0. # x-rotation drive strength
T_1 = 15.5 / MHz # single-photon decay time
k_a = 1. / T_1
n_th = 0.04 # equilibrium thermal |1>-state occupation

H_sys = hbar * (-K * qt.create(N)**2 * qt.destroy(N)**2 + e_2 * (qt.create(N)**2 + qt.destroy(N)**2)
         - 4. * K * np.abs(xi)**2 * qt.create(N) * qt.destroy(N) + e_x * qt.create(N) + np.conj(e_x) * qt.destroy(N))
```

```python
t = np.linspace(0., 0.8, 200) # 1 Î¼s
alpha = np.sqrt(e_2 / K)

psi0 = (qt.coherent(N, alpha) + qt.coherent(N, -alpha)) / np.sqrt(2. * (1. + np.exp(-2. * np.abs(alpha)**2)))
rho0 = psi0 * psi0.dag()

result = qt.mesolve(H_sys, rho0, t,
                    c_ops=[np.sqrt(k_a * (1. * n_th)) * qt.destroy(N), np.sqrt(k_a * n_th) * qt.create(N)],
                   options=qt.Options(store_states=True))

#print(np.array([x.full() for x in psi0.dag() * result.states * psi0]).flatten().real)
plt.plot(result.times, np.array([x.full() for x in psi0.dag() * result.states * psi0]).flatten().real)
plt.ylim(0., 1.)
```
